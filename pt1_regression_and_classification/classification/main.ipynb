### Preparando os modelos 

### Feature enginiering

```python
"""Normalização de variaveis numéricas"""
from sklearn.preprocessing import StandardScaler

numerical_features = ['age', 'education.num', 'capital.gain', 'capital.loss', 'hours.per.week']
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])

```

```python
"""OneHot encoding em variaveis categoricas não ordinárias"""
categorical_features = ['workclass', 'occupation', 'marital.status', 'relationship', 'race', 'sex', 'native.country']
df = pd.get_dummies(df, columns=categorical_features, drop_first=True)

for column in categorical_features:
    if (column.startswith(column)):
        if column in df.columns:
            df[column] = df[column].astype(int)
```

```python
"""Label encoding para dados categóricos ordinais"""
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['education'] = label_encoder.fit_transform(df['education'])
```

```python
X = df.loc[:, df.columns!="income"]
Y = df['income']
```

```python
"""Aplicando PCA em para reduzir dimensionalidade de features"""
from sklearn.decomposition import PCA

# Mantém 95% da variância
pca = PCA(n_components=0.95)
X[numerical_features] = pca.fit_transform(X[numerical_features])
```

### Treinamento de modelos

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, KFold, StratifiedKFold, cross_val_score, cross_val_predict, cross_validate
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, mean_squared_error

import re
```

```python
def rmse(test, predictions):
    rmse_value = np.sqrt(mean_squared_error(test, predictions))
    print(f"Root Mean Squared Error: {rmse_value}")

def display_scores(score):
    print(f"Scores {score}")
    print(f"Mean {score.mean()}")
    print(f"Standart deviation {score.std()}")
    print()

def display_feature_importance(classifier):
    feature_importances = pd.DataFrame(classifier.feature_importances_,
                                       index = df.columns.drop('income'),
                                        columns=['importance']).sort_values('importance', ascending=False)
    x = np.arange(len(df.columns.drop('income')))
    width = 0.25

    plt.figure(figsize=(12, 6))
    plt.bar(x - width, feature_importances['importance'], width, label='Hold-Out')
    plt.xticks(x, feature_importances.index, rotation=90)
    plt.ylabel("Importância dos Atributos")
    plt.title("Comparação da Importância de Atributos")
    plt.legend()
    plt.tight_layout()
    plt.show()


def display_metrics(test, predicitions):
    print("Acuracy:", accuracy_score(test, predicitions))
    print("Precision:", precision_score(test, predicitions))
    print("F1:", f1_score(test, predicitions))
    print("Recall:", recall_score(test, predicitions))

def display_confusion_matrix(test, predictions):
    conf_matrix = confusion_matrix(test, predictions)
    sns.heatmap(pd.DataFrame(conf_matrix), annot=True, cmap="Blues", fmt='g')
    plt.xlabel('Prediction')
    plt.ylabel('Test')
    plt.title('Confusion Matrix')
    plt.figure(figsize=(0.5, 0.5))
    plt.show()

    
def hold_out_cv(Classifier, X, Y):
    classifier = Classifier()
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25)
    classifier.fit(X_train, Y_train)
    predictions = classifier.predict(X_test)
    print("-- HOLD OUT --")
    rmse(Y_test, predictions)
    scores = cross_val_score(classifier, X, Y, scoring="neg_mean_squared_error", cv=10)
    score = np.sqrt(-scores)
    display_scores(score)
    display_metrics(Y_test, predictions)
    display_confusion_matrix(Y_test, predictions)
    display_feature_importance(classifier)

def k_fold_cv(Classifier, X, Y):
    print("-- K-FOLD --")
    classifier = Classifier()
    k_fold = KFold(n_splits=5, shuffle=True)
    pred = cross_val_score(classifier, X, Y, cv=k_fold, scoring="neg_mean_squared_error")
    predictions = cross_val_predict(classifier, X, Y, cv=k_fold)
    rmse(Y, predictions)
    rmse_value = np.sqrt(-pred)
    display_scores(rmse_value)
    display_metrics(Y, predictions)
    display_confusion_matrix(Y, predictions)

def stratified_k_fold_cv(Classifier, X, Y):
    print("-- STRATIFIED K-FOLD --")
    classifier = Classifier()
    sk_fold = StratifiedKFold(n_splits=5, shuffle=True)
    pred = cross_val_score(classifier, X, Y, cv=sk_fold, scoring="neg_mean_squared_error")
    predictions = cross_val_predict(classifier, X, Y, cv=sk_fold)
    rmse(Y, predictions)
    rmse_value = np.sqrt(-pred)
    display_scores(rmse_value)
    display_metrics(Y, predictions)
    display_confusion_matrix(Y, predictions)

def predict_and_analise(Classifier):
    print(re.sub(r'(?<!^)([A-Z])', r' \1', Classifier.__name__))

    hold_out_cv(Classifier, X, Y)  
    print("-"*25)
    k_fold_cv(Classifier, X, Y)
    print("-"*25)
    stratified_k_fold_cv(Classifier, X, Y)

```

### Decision Tree

```python
predict_and_analise(DecisionTreeClassifier)
```

```python
predict_and_analise(RandomForestClassifier)
```
